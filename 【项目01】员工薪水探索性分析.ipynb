```python
df_bar = df.groupby('员工职位')['基本工资'].mean().reset_index()
```

```python
df_bar.columns = ['员工职位','平均基本工资']
```

```python
df_bar.dropna(inplace=True)
```

```python
df_bar = df_bar[df_bar['平均基本工资']>0]
```

```python
df_bar['平均基本工资'] = round(df_bar['平均基本工资']/10000,0)
```

```python
df_bar = df_bar[df_bar['平均基本工资']>0]
```

```python
df_bar.sort_values(['平均基本工资'],ascending=False,inplace=True,ignore_index=True)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
```

```python
pye_bar = (
    Bar()
    .add_xaxis(
        list(df_bar['员工职位'])
    )
    .add_yaxis("平均基本工资", list(df_bar['平均基本工资']),color='grey')
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_="min", name="最小值"),
                opts.MarkLineItem(type_="max", name="最大值"),
                opts.MarkLineItem(type_="average", name="平均值"),
            ]
        ))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15)),
        title_opts=opts.TitleOpts(title="职位的平均基本工资"),
        datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")]
    ))
```

```python
pye_bar.render_notebook()
```

```python
df_bar2 = df_bar.groupby('平均基本工资').count().reset_index()
```

```python
df_bar2.columns = ['平均基本工资','职位数量']
```

```python
df_bar2.head()
```

```python
pye_bar2 = (
    Bar()
    .add_xaxis(
        list(df_bar2['平均基本工资'])
    )
    .add_yaxis("职位数量", list(df_bar2['职位数量']),color='#66ccff')
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markline_opts=opts.MarkLineOpts(
            data=[
                opts.MarkLineItem(type_="min", name="最小值"),
                opts.MarkLineItem(type_="max", name="最大值"),
                opts.MarkLineItem(type_="average", name="平均值"),
            ]
        ))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15)),
        title_opts=opts.TitleOpts(title="平均基本工资的职位数量"),
        datazoom_opts=[opts.DataZoomOpts(), opts.DataZoomOpts(type_="inside")]
    ))
```

```python
pye_bar2.render_notebook()
```
